# Question -2

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.bcci.tv/")
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[5]/div[2]/div/div/div[2]/div/div[2]/div/div/div/div[1]/a")
search.click()  
```

```python
Series=[]
Place=[]
date=[]
Time=[]
```

```python
series_tags =driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]')
for i in series_tags:
    title=i.text
    Series.append(title)
```

```python
Series
```

```python
place_tags =driver.find_elements(By.XPATH,'//div[@class="match-venue ng-scope"]')
for i in place_tags:
    title=i.text
    Place.append(title)
```

```python
Place
```

```python
date_tags =driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]')
for i in date_tags:
    title=i.text
    date.append(title)
```

```python
date
```

```python
time_tags =driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]')
for i in time_tags:
    title=i.text
    Time.append(title)
```

```python
Time
```

```python
print(len(Series),len(Place),len(date),len(Time))
```

```python
df = pd.DataFrame({'Series':Series,'Place':Place,'Date':date,"Time":Time})
df
```
